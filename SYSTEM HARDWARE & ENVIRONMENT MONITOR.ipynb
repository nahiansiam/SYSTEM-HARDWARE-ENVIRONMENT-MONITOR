```python
import platform
import psutil
import subprocess
import socket
from datetime import datetime
from cpuinfo import get_cpu_info


# ============================
# HEADER UI
# ============================
def line():
    print("═" * 70)


def title(text):
    print(f"\n🔷 {text}")
    print("-" * 70)


# ============================
# SYSTEM INFO
# ============================
def get_windows_version():
    return f"{platform.system()} {platform.release()}"


def get_python_version():
    return platform.python_version()


def get_arch():
    return platform.architecture()[0]


def get_boot_time():
    boot = psutil.boot_time()
    return datetime.fromtimestamp(boot).strftime("%Y-%m-%d %H:%M:%S")


# ============================
# CPU
# ============================
def get_cpu():
    info = get_cpu_info()
    name = info.get("brand_raw", "Unknown CPU")

    cores = psutil.cpu_count(logical=False)
    threads = psutil.cpu_count(logical=True)
    usage = psutil.cpu_percent(interval=1)

    freq = psutil.cpu_freq()
    freq_ghz = round(freq.current / 1000, 2) if freq else 0

    return name, cores, threads, freq_ghz, usage


# ============================
# GPU (NAME + USAGE ONLY)
# ============================
def get_gpu_name():
    try:
        result = subprocess.check_output(
            "wmic path win32_VideoController get name",
            shell=True
        ).decode(errors="ignore").split("\n")

        for line in result:
            line = line.strip()
            if line and "Name" not in line:
                return line
    except:
        pass

    return "Unknown GPU"


def get_gpu_usage():
    try:
        result = subprocess.check_output(
            "wmic path Win32_PerfFormattedData_GPUPerformanceCounters_GPUEngine get UtilizationPercentage",
            shell=True
        ).decode(errors="ignore").split("\n")

        values = [int(x.strip()) for x in result if x.strip().isdigit()]
        return max(values) if values else 0
    except:
        return 0


# ============================
# RAM
# ============================
def get_ram():
    r = psutil.virtual_memory()
    return (
        round(r.total / (1024**3), 2),
        round(r.available / (1024**3), 2),
        r.percent
    )


# ============================
# STORAGE
# ============================
def get_disk():
    d = psutil.disk_usage("C:\\")
    return (
        round(d.total / (1024**3), 2),
        round(d.used / (1024**3), 2),
        round(d.free / (1024**3), 2),
        d.percent
    )


# ============================
# NETWORK
# ============================
def get_network():
    host = socket.gethostname()
    ip = socket.gethostbyname(host)
    return host, ip


def internet_status():
    try:
        socket.create_connection(("8.8.8.8", 53), timeout=2)
        return "🟢 Connected"
    except:
        return "🔴 Disconnected"


# ============================
# MAIN DASHBOARD
# ============================
def main():
    line()
    print("🖥️  SYSTEM HARDWARE & ENVIRONMENT MONITOR")
    line()

    print(f"🪟 OS            : {get_windows_version()}")
    print(f"🐍 Python        : {get_python_version()}")
    print(f"⚙️ Architecture   : {get_arch()}")
    print(f"⏱️ Boot Time     : {get_boot_time()}")

    # CPU
    cpu = get_cpu()
    title("CPU")
    print(f"🧠 Name          : {cpu[0]}")
    print(f"🧩 Cores         : {cpu[1]}")
    print(f"🔩 Threads       : {cpu[2]}")
    print(f"⚡ Frequency     : {cpu[3]} GHz")
    print(f"📊 Usage         : {cpu[4]} %")

    # GPU
    title("GPU")
    print(f"🎮 Name          : {get_gpu_name()}")
    print(f"📊 Usage         : {get_gpu_usage()} %")

    # RAM
    ram = get_ram()
    title("RAM")
    print(f"📦 Total         : {ram[0]} GB")
    print(f"📥 Available     : {ram[1]} GB")
    print(f"📊 Usage         : {ram[2]} %")

    # STORAGE
    disk = get_disk()
    title("STORAGE (C Drive)")
    print(f"💾 Total         : {disk[0]} GB")
    print(f"📂 Used          : {disk[1]} GB")
    print(f"🗄️ Free          : {disk[2]} GB")
    print(f"📊 Usage         : {disk[3]} %")

    # NETWORK
    net = get_network()
    title("NETWORK")
    print(f"🏠 Hostname      : {net[0]}")
    print(f"🌐 Local IP      : {net[1]}")
    print(f"📡 Internet      : {internet_status()}")

    line()
    print("✅ SCAN COMPLETED SUCCESSFULLY")
    line()

    # ============================
    # CREDITS
    # ============================
    print("\n👨‍💻 Created by : Nahian Kabir Siam")
    print("🎓 Role       : Computer Science & Engineering (CSE)")
    print("⚙️ Version    : Stable Windows 10/11 Edition")
    line()


if __name__ == "__main__":
    main()
```
